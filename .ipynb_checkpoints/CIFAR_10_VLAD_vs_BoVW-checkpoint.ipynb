## Keypoint Detection
- SURF used to detect keypoints
- Each keypoint has a dimensionality of 64
- Codebook built on keypoint descriptors of training images, with sizes of 16, 64 and 256

```python
image = cv2.resize(train_data[10], (150, 150), interpolation=cv2.INTER_LINEAR)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 

surf = cv2.xfeatures2d.SURF_create()
kp, des = surf.detectAndCompute(gray,None)

copy = cv2.drawKeypoints(gray,kp, outImage = None)
plt.imshow(copy[:, :, [2, 1, 0]])
```

```python
surf_keypoints = []

for image in train_data :
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    image = cv2.resize(image, (150, 150), interpolation=cv2.INTER_LINEAR)
    surf = cv2.xfeatures2d.SURF_create()
    kp, descriptors = surf.detectAndCompute(image,None)
    surf_keypoints.append(descriptors)


surf_keypoints = np.concatenate(surf_keypoints, axis=0)
k = [16, 64, 256]
mbk = []
for i in k:
    mbk.append(MiniBatchKMeans(init='k-means++', n_clusters=i,max_iter=1000, batch_size=batch_size,
                  n_init=10, max_no_improvement=10, verbose=0).fit(surf_keypoints))
# label=mbk.labels_
# centroid=mbk.cluster_centers_
with open("codebook.pickle",'wb')as pk:
    pickle.dump(mbk,pk)
```

## Feature Extraction

### Bag of Visual Words (BoVW)

```python
def calculate_histogram(images, model):

    feature_vectors=[]

    for image in images :

        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (150, 150), interpolation=cv2.INTER_LINEAR)
        #SURF extraction
        surf = cv2.xfeatures2d.SURF_create()
        kp, descriptors = surf.detectAndCompute(image,None)
        #classification of all descriptors in the model
        predict_kmeans = model.predict(descriptors)
        #calculates the histogram
        hist, bin_edges = np.histogram(predict_kmeans, bins = 10)
        #histogram is the feature vector
        feature_vectors.append(hist)

    feature_vectors=np.asarray(feature_vectors)

    return np.array(feature_vectors)
```

```python
x_feat_train = []
x_feat_test = []

for i in range(3):
    x_feat_train.append(calculate_histogram(train_data, mbk[i]))
    x_feat_test.append(calculate_histogram(test_data, mbk[i]))
```

#### Model Evaluation with SVM - BoVW

```python
clf = svm.LinearSVC(multi_class='ovr')
for i in range(3):
    clf.fit(x_feat_train[i], train_labels)
    y_pred = clf.predict(x_feat_test[i])
    print(accuracy_score(y_pred, test_labels))
```

### Vector of Locally Aggregated Descriptors (VLAD)

```python
def VLAD(X,visualDictionary=mbk):

    predictedLabels = visualDictionary.predict(X)
    centers = visualDictionary.cluster_centers_
    labels=visualDictionary.labels_
    k=visualDictionary.n_clusters
   
    m,d = X.shape
    V=np.zeros([k,d])
    #computing the differences

    # for all the clusters (visual words)
    for i in range(k):
        # if there is at least one descriptor in that cluster
        if np.sum(predictedLabels==i)>0:
            # add the diferences
            V[i]=np.sum(X[predictedLabels==i,:]-centers[i],axis=0)
    

    V = V.flatten()
    # power normalization, also called signed square-rooting normalization
    V = np.sign(V)*np.sqrt(np.abs(V))

    # L2 normalization

    V = V/np.sqrt(np.dot(V,V))
    return V
```

```python
def getVLADDescriptors(images, model):
    descriptors=list()
#     idImage =list()
    surf = cv2.xfeatures2d.SURF_create()
    for im in images:
        image = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (150, 150), interpolation=cv2.INTER_LINEAR)
        kp,des = surf.detectAndCompute(image, None)
        if des.any():
            v=VLAD(des, model)
            descriptors.append(v)
#             idImage.append(imagePath)
                    
    #list to array    
    descriptors = np.asarray(descriptors)
#     return descriptors, idImage
    return descriptors
```

```python
x_feat_train = []
x_feat_test = []
for i in range(3):
    x_feat_train.append(getVLADDescriptors(train_data, mbk[i]))
    x_feat_test.append(getVLADDescriptors(test_data, mbk[i]))
```

#### Model Evaluation - VLAD

```python
clf = svm.LinearSVC(multi_class='ovr')
clf = svm.LinearSVC(multi_class='ovr')
for i in range(3):
    clf.fit(x_feat_train[i], train_labels)
    y_pred = clf.predict(x_feat_test[i])
    if i == 1:
        cm = ConfusionMatrix(clf)
        cm.fit(x_feat_train[1], train_labels)
        cm.score(x_feat_test[1], test_labels)
    print(accuracy_score(y_pred, test_labels))
```
